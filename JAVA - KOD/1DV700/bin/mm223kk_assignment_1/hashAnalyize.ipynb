```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
result = np.loadtxt('words.csv', delimiter=',') # Get all results.
X = result[:,0:1] # Reoder the columns.
y = result[:,1:]

plt.figure(figsize=(10, 6))
plt.plot(X, y, "c-", label='Collions')
plt.title('Collions - Random Words')
plt.xlabel('X - Words')
plt.ylabel('Y - Collioons')
plt.legend()
plt.show()
```
