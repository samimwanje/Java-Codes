```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
sum = np.loadtxt('unionFind.csv', delimiter=',') # Get all results.
X = sum[:,0:1] # Reoder the columns.
y = sum[:,1:]

sum_si = np.loadtxt('unionFind_q.csv', delimiter=',') # Get all results.
X_q = sum_si[:,0:1] # Reoder the columns.
y_q = sum_si[:,1:]

plt.figure(figsize=(10, 6))
plt.plot(X, y, "c-", label='unionFind')
plt.plot(X_q, y_q, "k-", label='unionFind_q')
plt.title('unionFind vs unionFind_q')
plt.xlabel('Number of unions)')
plt.ylabel('time (seconds)')
plt.legend()
plt.show()
```

```python
sum = np.loadtxt('unionFind100k.csv', delimiter=',') # Get all results.
X = sum[:,0:1] # Reoder the columns.
y = sum[:,1:]

sum_si = np.loadtxt('unionFind_q100k.csv', delimiter=',') # Get all results.
X_q = sum_si[:,0:1] # Reoder the columns.
y_q = sum_si[:,1:]

plt.figure(figsize=(10, 6))
plt.plot(X, y, "c-", label='unionFind')
plt.plot(X_q, y_q, "k-", label='unionFind_q')
plt.title('unionFind vs unionFind_q')
plt.xlabel('Number of unions)')
plt.ylabel('time (seconds)')
plt.legend()
plt.show()
```

This exercise was intresting since theoretically union find should be slower than quick union find. Union find has to iterate thorugh all the elements before a union, while union find uses root(x), which also iterate through all the elements but backwards. 

The result seen above shows that union find is much faster then quck union find when it comes to Java. In the examples from the lesson quick UnionFind was faster than UnionFind, but that was on a python example.

When it comes to java we first need to go through what the diffrence between these two are. First of a the function root(x) is called for the QUF. In the basic UF the program iterates through the array using a for-loop.  

The main difference here is that when the function root(x) is called a while loop is used to find the root of x. The conclusion here should therefore be that a while loop iteration is much slower than a for-loop iteration in Java. This may be the reason for the results above.
