```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
sum = np.loadtxt('3sum.csv', delimiter=',') # Get all results.
X = sum[:,0:1] # Reoder the columns.
y = sum[:,1:]

sum_si = np.loadtxt('3sum_si.csv', delimiter=',') # Get all results.
X_si = sum_si[:,0:1] # Reoder the columns.
y_si = sum_si[:,1:]

plt.figure(figsize=(10, 6))
plt.plot(X, y, "c-", label='3sum Brute-force')
plt.plot(X_si, y_si, "k-", label='3sum_si')
plt.title('3sum vs 3sum_si')
plt.xlabel('length(array)')
plt.ylabel('time (seconds)')
plt.legend()
plt.show()
```

Above we can see the result of two implementations of 3sum Brute-force and 3sum_si.

Theoretically 3sum_si should be faster than 3sum. Since we are not interating over the whole list every time.

And there is no final check if (i,j) or (j,i) is already included.

As we can see in the result above 3sum_si becomes faster than 3sum-brute-force after an array of the length 300+.

This may occur because after an array of the length 100 + it takes significantly longer time to iterate over each element in the array.

With this said brute force method also need to check each element in the array. Even the chked one. 

This can be more time consuming in the end if we compare it to 3sum_si which uses i+1 and i+2 to skip the previous elements in each loop.
