```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
result = np.loadtxt('percolation_nxn.csv', delimiter=',') # Get all results.
X = result[:,0:1] # Reoder the columns.
y = result[:,1:]

plt.figure(figsize=(10, 6))
plt.plot(X, y, "c-", label='Open gates')
plt.title('Percolation - NxN')
plt.xlabel('NxN')
plt.ylabel('Open gates')
plt.legend()
plt.show()
```

The graph above  shows the gates needed to be open in order for a system wih NxN to perlocate.

After som runs we ce can see that the result of a NxN perlocatation system has the growth "N^2".

For this particular graph the growth was y = 0,59 * x^2. 

The number 0,59 can be seen as a number which is the percolation threshold.

This means that when more than 60 % of the gates  in a NxN system are open the system will most likely perclocate.

A number of open gates bellow 0,59 (60 %) would most likely not perclocate the system.

```python
result = np.loadtxt('percolation_10x10.csv', delimiter=',') # Get all results.
X = result[:,0:1] # Reoder the columns.
y = result[:,1:]

plt.figure(figsize=(10, 6))
plt.plot(X, y, "c-", label='Open gates')
plt.title('Percolation - 10x10')
plt.xlabel('Runs - (x)')
plt.ylabel('Open gates')
plt.legend()
plt.show()
```

In the above graph we can see the measuring results when using a static 10x10 system. This was measured 1000 times (stamps), and each stamps open gates was saved.

In the results above we can see that the "thickest" stamps of "Open gates" are in the middle just under 60 and above 55. This proves that then if open gates are in the range of of  p > 55, the system will most likely perclocate. 

A more exakt  value was measured using the mean of the 1000 stamps. That value was around 59.


The avarage p* is 59 after around 1000 stamps. This means that when there are more than 59 random gates open the system is most likely to perlocate.

The threshold is about 0.59, which means around 60 % of the gates needs to be open in order for the program to percolate.
